## Q1 What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a Python library for creating static, animated, and interactive visualizations in Python. It is one of the most widely used visualization libraries in the Python community.

Matplotlib is used for creating various types of plots, such as line plots, scatter plots, bar plots, histograms, and many more. It is an essential tool for data analysis and visualization, as it allows users to present data in a visually appealing manner and gain insights into the data.

 plots that can be plotted using the Pyplot module of Matplotlib are:

1. Line Plot - A plot of lines connecting data points in the order they are provided.

2. Scatter Plot - A plot of individual data points, where the x and y coordinates are specified.

3. Bar Plot - A plot of rectangular bars representing the data.

4. Histogram - A plot that displays the distribution of a dataset.

5. Pie Chart - A plot that displays the proportion of each category in a dataset as a slice of a pie.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.


A scatter plot is a type of plot that displays values for two variables as points on a two-dimensional coordinate plane. Each point on the plot represents the value of the two variables for a single data point.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Small Demonstration of Scatter plot")
plt.figure(figsize=(10,10))
plt.show()
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```


A bar plot is a visualization that displays rectangular bars with lengths proportional to the values they represent. It is also known as a bar chart. A bar plot is typically used to compare the values of different categories or groups. The categories or groups are usually represented on the x-axis, while the values are represented on the y-axis.

Bar plots are useful for presenting categorical data because they allow easy comparisons between the groups. They are commonly used in business, finance, and economics to display data such as sales figures, stock prices, or financial performance. 

Bar plots can be used to represent different types of data, including frequency counts, percentages, and numerical values. They are easy to create and interpret, making them a popular choice for data visualization.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.barh(company,profit,height=0.8)
plt.ylabel("Company Names")
plt.xlabel("Profits")
plt.title("Horizontal Bar plot")
plt.show()
```

## Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

A box plot is a graphical representation of the statistical summary of a set of data through their quartiles. The plot consists of a box and whiskers, where the box represents the interquartile range (IQR) containing the middle 50% of the data, the line inside the box represents the median, and the whiskers extend to the minimum and maximum values that fall within 1.5 times the IQR from the lower and upper quartiles, respectively. Any data points beyond the whiskers are considered outliers and are plotted individually as dots or asterisks.

Box plots are used to visualize the distribution of a set of data and to identify outliers and extreme values. They can also be used to compare the distribution of multiple sets of data side by side.

```python
box1
```

```python
box2
```

```python
import matplotlib.pyplot as plt
fig =plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.boxplot(box1)
plt.title("Box-1")

plt.subplot(1,2,2)
plt.boxplot(box2)
plt.title("Box-2")

fig.suptitle(" Box plot")
plt.show()
```
