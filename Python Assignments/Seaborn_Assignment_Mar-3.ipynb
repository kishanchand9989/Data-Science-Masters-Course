### Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Seaborn is a popular data visualization library built on top of Matplotlib. It provides a high-level interface for creating informative and visually appealing statistical graphics. 
Here are five commonly used plots in Seaborn along with their use cases:

Bar Plot 

A bar plot represents categorical data with rectangular bars whose lengths indicate the values they represent. It is useful for comparing values across different categories.
   Use cases: Visualize counts or frequencies of categorical variables, compare values between different groups or categories.

Histogram

A histogram displays the distribution of a continuous variable by dividing the data into bins and showing the count or density of observations in each bin. It provides insights into the underlying data distribution.
   Use cases: Understand the distribution and range of a variable, identify skewness or outliers, analyze data spread.

Scatter Plot

A scatter plot shows the relationship between two variables as individual points. Each point represents an observation, and the position of the point represents the values of the two variables.
   Use cases: Explore the correlation or association between two continuous variables, identify patterns or clusters, visualize trends.

Heatmap 

A heatmap visualizes tabular data using colors to represent the values in a matrix. It provides a compact and intuitive way to understand relationships between variables in a dataset.
   Use cases: Display correlation matrices, explore patterns or relationships in multi-dimensional data, highlight high or low values in a matrix.

Box Plot

A box plot, also known as a box-and-whisker plot, displays the summary statistics of a continuous variable. It shows the median, quartiles, and possible outliers in a dataset.
   Use cases: Visualize the distribution of a continuous variable, compare summary statistics between different categories or groups, detect outliers.



## Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions. 
## Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
df =sns.load_dataset("fmri")
df.head()
```

```python
sns.lineplot(x=df["timepoint"],y=df["signal"],hue=df['event'])
```

```python
sns.lineplot(x=df["timepoint"],y=df["signal"],hue=df['region'])
```

## Q 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt
df5 =sns.load_dataset("iris")
sns.pairplot(data=df5,hue="species")
plt.show
```

## Q6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import seaborn as sns

# Load the "flights" dataset
flights = sns.load_dataset("flights")

# Reshape the dataset for heatmap
flights_pivot = flights.pivot("month", "year", "passengers")

# Create a heatmap
sns.heatmap(data=flights_pivot, annot=True, fmt="d", cmap="YlGnBu")

# Set plot title and labels
plt.title("Passenger Traffic")
plt.xlabel("Year")
plt.ylabel("Month")

# Show the plot
plt.show()
```
